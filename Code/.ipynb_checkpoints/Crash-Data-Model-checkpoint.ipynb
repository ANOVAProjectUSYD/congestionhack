Generate a day for the whole time frame. If it's not there, then generate a new one.

```python
from datetime import timedelta

start = datetime.strptime('1-Jan-12', '%d-%b-%y')
end = datetime.strptime('28-Oct-17', '%d-%b-%y')

delta = end - start         # timedelta

datetimes = [] # Stores our datetimes

for i in range(delta.days + 1):
    new_date = (start + timedelta(days=i))
    datetimes.append(new_date)
```

```python
final_df = pd.DataFrame({'Date': datetimes}) # Final pandas dataframe
```

```python
dates_of_crashes = list(sydney_df.modified_date)
```

```python
final_df['Crash'] = final_df[final_df.Date.isin(dates_of_crashes)]
```

```python
final_df.Crash.loc[~final_df.Crash.isnull()] = 1 
final_df.Crash.replace(np.NaN, 0, inplace=True)
```

## Adding in seasons

```python
# December to February is summer; 
# March to May is autumn; 
# June to August is winter; 
# and September to November is spring.

autumn = range(61, 152)
winter = range(152, 243)
spring = range(243, 334)
# summer = everything else
```

```python
if final_df.Date[0].day in autumn:
    print("No")
else:
    print("Yes")
```

```python
final_df['Autumn'] = final_df['Date'].dt.dayofyear.isin(autumn)
final_df['Winter'] = final_df['Date'].dt.dayofyear.isin(winter)
final_df['Spring'] = final_df['Date'].dt.dayofyear.isin(spring)
# summer = everything else

final_df.Autumn.replace(False, 0, inplace=True)
final_df.Autumn.replace(True, 1, inplace=True)
final_df.Winter.replace(False, 0, inplace=True)
final_df.Winter.replace(True, 1, inplace=True)
final_df.Spring.replace(False, 0, inplace=True)
final_df.Spring.replace(True, 1, inplace=True)

final_df['Autumn'] = final_df.Autumn.astype(int)
final_df['Winter'] = final_df.Winter.astype(int)
final_df['Spring'] = final_df.Spring.astype(int)
```

## Check if weekday

```python
final_df['Weekend'] = final_df['Date'].dt.dayofweek >= 5 
final_df.Weekend.replace(False, 0, inplace=True)
final_df.Weekend.replace(True, 1, inplace=True)
final_df['Weekend'] = final_df.Weekend.astype(int)
```

## Model

```python
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
final_df.to_csv("final_datasets.csv", index=False)
```

```python
datasets = pd.read_csv("final_datasets.csv")
```

```python
X = datasets
X.pop("Date")
y = X.pop("Crash")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#Computes accuracy and confusion matrices
#Not normalising features
def predictingModel(model):
    
    accuracy = []
    confusion_matrices = []
    
    for y_type in y_train:
        model.fit(X_train, (y_train))
        pred = model.predict(X_test)
        confusion_matrices.append(confusion_matrix(y_test,pred))
        accuracy.append(accuracy_score(y_test[y_type],pred))
    return accuracy, confusion_matrices
```

```python
#Logistic regression
LogReg = LogisticRegression(random_state=450411920)
LogReg.fit(X_train, y_train)
```

```python
y_pred = LogReg.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
confusion_matrix(y_pred, y_test)
```
