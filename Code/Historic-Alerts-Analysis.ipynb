```python
import pandas as pd
import numpy as np
import datetime
```

```python
data = pd.read_csv('../Datasets/alerts_2017_jun_to_dec.csv')
data['active_from_dttm'] = pd.to_datetime(data['active_from_dttm'])
data['active_to_dttm'] = pd.to_datetime(data['active_to_dttm'])
```

```python
df = data.set_index(pd.DatetimeIndex(data['active_from_dttm']))
```

```python
data_weekends = df[df.index.dayofweek>5]
```

```python
data_weekends['alert_description'] = data_weekends.alert_description.str.lower()
```
