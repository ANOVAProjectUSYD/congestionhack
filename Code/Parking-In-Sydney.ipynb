# Paid Street Parking in Sydney

```python
#load json object
with open('../Datasets/paid_street_parking.json') as f:
    d = json.load(f)
    f.close()
```

```python
import csv
def initial_write():
    '''Create and write first entry in file.'''
    csv_file = "paid_street_parking" + '.csv'
    write_file = open(csv_file, 'w')
    writer = csv.writer(write_file)
    list_of_keys = []
    for x in d[0]['attributes'].keys():
        list_of_keys.append(x)
    for x in d[0]['geometry'].keys():
        list_of_keys.append(x)
    writer.writerow(list_of_keys)
    write_file.close()
```

```python
initial_write()
```

```python
def write_to_csv():
    '''Writes current json entry into csv'''
    csv_file = "paid_street_parking" + '.csv'
    write_file = open(csv_file, 'a')
    writer = csv.writer(write_file)
    list_of_values = []
    for i in range(0, len(d)):
        list_of_values.clear()
        for x in d[i]['attributes'].values():
            list_of_values.append(x)
        for x in d[i]['geometry'].values():
            list_of_values.append(x)
        writer.writerow(list_of_values)
    write_file.close()
```

```python
write_to_csv()
```
